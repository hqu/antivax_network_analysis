```python
# prompt: Upload VAX_G1.csv and load it into a table

from google.colab import files
import pandas as pd

# Upload the file
uploaded = files.upload()

# Assuming the file is named 'VAX_G1.csv'
if 'VAX_G1.csv' in uploaded:
  df = pd.read_csv('VAX_G1.csv')
  df.head()
else:
  print("File 'VAX_G1.csv' not found in uploaded files.")
```

```python
df
```

```python
# prompt: filter df table to a new df_network table with only the Twitter Users and Mention columns

df_network = df[['Twitter Users', 'Mention']]

df_network
```

```python
# prompt: delete all the rows in which Mention value is NaN

df_network = df_network.dropna(subset=['Mention'])

df_network
```

```python
# prompt: In the Mention column, remove all the "@" characters
df_network['Mention'] = df_network['Mention'].str.replace('@', '')
df_network
```

```python
# prompt: From the Mention column, create a frequency table df_frequency
# that contains Mention and the number of time that Mention occurs sorted in descencding order
df_frequency = df_network['Mention'].value_counts().rename_axis('Mention').reset_index(name='Frequency')
df_frequency_sorted = df_frequency.sort_values(by='Frequency', ascending=False)
df_frequency_sorted[:100]
```

```python
# prompt: Upload VAX_G2.csv and load it into a table

from google.colab import files
import pandas as pd

uploaded = files.upload()

# Assuming the file is named 'VAX_G2.csv'
if 'VAX_G2.csv' in uploaded:
  df_g2 = pd.read_csv('VAX_G2.csv')
  df_g2.head()
else:
  print("File 'VAX_G2.csv' not found in uploaded files.")
```

```python
# prompt: delete all the rows in df_network_g2 in which Mention value is NaN

df_network_g2 = df_network_g2.dropna(subset=['Mention'])
df_network_g2
```

```python
# prompt: From the Mention column in df_network_g2, create a frequency table df_frequency_g2 that contains Mention and the number of time that Mention occurs sorted in descencding order, then display the top 100.

df_frequency_g2 = df_network_g2['Mention'].value_counts().rename_axis('Mention').reset_index(name='Frequency')
df_frequency_g2_sorted = df_frequency_g2.sort_values(by='Frequency', ascending=False)
df_frequency_g2_sorted[:100]
```
